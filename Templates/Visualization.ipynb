```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
```

```python
df = pd.DataFrame(np.random.randn(200, 5), columns=["A","B","C","D","E"])
```

## Plot

```python
df.plot(x="A", y="B", kind='line', ax=None,
        subplots=False, sharex=None, sharey=False,
        layout=None, figsize=None, use_index=True,
        title=None, grid=None, legend=True, style=None,
        logx=False, logy=False, loglog=False, xticks=None,
        yticks=None, xlim=None, ylim=None, rot=None, fontsize=None,
        colormap=None, table=False, yerr=None, xerr=None, secondary_y=False, sort_columns=False)
plt.show()
```

Different plot can be provided as the **kind** keyword argument to plot(). These include:

```python
# 'line'    : line plot (default)
# 'bar'     : vertical bar plot
# 'barh'    : horizontal bar plot
# 'hist'    : histogram
# 'box'     : boxplot
# 'kde'     : Kernel Density Estimation plot
# 'density' : same as 'kde'
# 'area'    : area plot
# 'pie'     : pie plot
# 'scatter' : scatter plot
# 'hexbin'  : hexbin plot
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=300, facecolor='w', edgecolor='k')
df.ix[0].plot(kind='kde')
plt.axhline(0, color='k')
plt.show()
```

## Joint Plot

```python
#kind = "scatter","reg","resid","kde","hex"
sns.jointplot(x = df["A"],
              y = df["B"],
           kind = 'scatter')
plt.show()
```

```python
sns.jointplot(x = df["A"],
              y = df["B"],
           kind = "kde")
plt.show()
```

```python
sns.jointplot(x = df["A"],
              y = df["B"],
           kind = "hex")
plt.show()
```

## Bar Plot

```python
plt.figure(num=None, figsize=(28, 26), dpi=300, facecolor='w', edgecolor='k')
df.plot.bar()
plt.show()
```

```python
plt.figure()
df.plot.bar(stacked=True)
plt.show()
```

```python
df.plot.barh(stacked=True)
plt.show()
```

```python
plt.figure()
df.plot.hist(alpha=0.7)
plt.show()
```

```python
plt.figure()
df.plot.hist(stacked=True, bins=20)
plt.show()
```

```python
df.diff().hist(color='k', alpha=0.5, bins=50)
plt.show()
```

## Box Plot

```python
df.plot.box()
plt.show()
```

```python
color = dict(boxes = 'DarkGreen',
          whiskers = 'DarkOrange',
           medians = 'DarkBlue',
              caps = 'Gray')
df.plot.box(color=color, sym='r+')
plt.show()
```

## Correlation Plot

```python
sns.corrplot(df) # compute and plot the pair-wise correlations
plt.show()
```

## Violin Plot

```python
#bw    : "scott", "silverman", float
#scale : "area", "count", "width"
#inner : "box", "quartile", "point", "stick", None

sns.violinplot(df, hue=None, data=None, order=None, hue_order=None, bw='silverman', cut=2, scale='area',
               scale_hue=True, gridsize=100, width=0.8, inner='box', split=False, orient=None, linewidth=None,
               color=None, palette=None, saturation=0.75, ax=None)
plt.show()
```
